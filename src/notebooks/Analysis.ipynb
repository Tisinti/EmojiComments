```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
```

```python
data = pd.read_parquet('../../data/probing/embeds.train.parquet')
```

```python
embeds = PCA(n_components=3).fit_transform(np.stack(data.embeds.tolist()))
data[['x', 'y', 'z']] = [(emb[0], emb[1], emb[2]) for emb in embeds]
```

```python
import plotly.graph_objects as go

def plot3d(data: pd.DataFrame) -> None:
    PLOT = go.Figure()
    labels = list(data['Label'].unique())
    labels.remove('joy')
    
    for label in labels:
        
        PLOT.add_trace(go.Scatter3d(x = data[data['Label'] == label]['x'],
                                    y = data[data['Label'] == label]['y'],
                                    z = data[data['Label'] == label]['z'],
                                    mode = 'markers', marker_size = 4, marker_line_width = 1,
                                    name = str(label)))


    PLOT.update_layout(width = 800, height = 800, autosize = True, showlegend = True,
                    scene = dict(xaxis=dict(title = 'x', titlefont_color = 'black'),
                                    yaxis=dict(title = 'y', titlefont_color = 'black'),
                                    zaxis=dict(title = 'z', titlefont_color = 'black')),
                    font = dict(family = "Gilroy", color  = 'black', size = 14))
    
    return PLOT
```

```python
plot3d(data)
```

```python
from sklearn.manifold import TSNE
embeds = TSNE(n_components=3, learning_rate='auto',init='random', perplexity=10).fit_transform(np.stack(data.embeds.tolist()))
data[['x', 'y', 'z']] = [(emb[0], emb[1], emb[2]) for emb in embeds]
```

```python
plot3d(data)
```
